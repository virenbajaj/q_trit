```python
import numpy as np
import matplotlib.pyplot as plt
from util import *
```

```python
d = load_obj("optVarDict")
```

```python
for bList in d.values():
    for node in bList:
        node.calcDist()
        node.calcEVals()
```

```python
save_obj(d,"optVarDict")
```

```python
cRange = np.linspace(-.4,.9,13)
for c in cRange:
    c = round(c,4)
    cStr = str(c)
    key = hash(cStr)
    cNodes = d[key]
    bVals = [node.b for node in cNodes]
    dVals = [node.dist for node in cNodes]
    #begin figure and plot
    fig = plt.figure()
    bRange = calcBRange(c)
    (bMin,bMax) = bRange[0]-1, bRange[-1]+1
    #make x,y axis
    # plt.plot((-1.,0.5),(0,0),'k--')
    # plt.plot((0,0),(-1.0,1.0),'k--')
    plt.xlabel('b', fontsize=15)
    plt.ylabel('dist', fontsize=15)
    plt.xlim(bMin,bMax)
    dMax = max(dVals)+1
    plt.ylim(0,dMax)
    plt.plot(bVals,dVals,"o")
    #title
    title = 'dist vs b for c = {0}'.format(cStr)
    print title              
    fig.suptitle(title, fontsize=20)
    # plt.show()

    title1 = "plots/dVsb1_{}.png".format(cStr) # save file to subdirectory called plots
    fig.savefig(title1)
```
